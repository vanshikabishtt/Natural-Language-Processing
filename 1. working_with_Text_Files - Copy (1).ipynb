## Working with Text Files

```python
name="Fred"
print(f'His name is {name}.')
```

Pass !r to get the string representation

```python
print(f'His name is {name!r}')
```

```python
d={'a':123,'b':456}
print(f'Addrress:{d['a']} Main Street')
```

```python
import pandas as pd
a={'author':['Twain','Feynman','Hamilton'],'Topic':['Rafting','Physics','Mythology'],'Pages':[601,95,144]}
b=pd.DataFrame(a)
for i in b:
    print(f'{b}')
    break
```

```python
library = [('Author','Topic','Pages'),('Twain','Rafting',601),('Feynman','Physics',95),('Hamilton','Mythology',144)]
for books in library:
    print(f'{books[0]:{10}} {books[1]:{8}} {books[2]:{7}}')
```

```python
library = [('Author','Topic','Pages'),('Twain','Rafting',601),('Feynman','Physics',95),('Hamilton','Mythology',144)]
for books in library:
    print(f'{books[0]:{15}} {books[1]:{15}} {books[2]:{4}}')
```

```python
for books in library:
    print(f'{books[0]:{15}} {books[1]:{15}} {books[2]:.>{8}}')
```

```python
for books in library:
    print(f'{books[0]:{15}} {books[1]:{15}} {books[2]:.>{10}}')
```

## Creating a file with Python

```python
%%writefile test1.txt
Hello, this is a quick test file.
This is the second line of the file.
```

```python
#open the text.txt file we created earlier
my_file=open('test.txt1')
my_file
```

#### .read() and .seek()

```python
#we can now read the file
my_file.read()
```

```python
# if we run the .read() command it will not give any ouput for this we have to run .seek() command to again read the file
my_file.seek(0)
```

#### You can read a file line by line using the readlines method.

```python
#Readlines returns a list of the lines in the file
my_file.seek(0)
my_file.readlines()
```

```python
#Whenever we will open a file, it's mandatory to close that file
my_file.close()
```

### Write in a file
##### the argument 'w' is used to write over a file. 

```python
#Add a second argument to the function, 'w' which stand for write.
#Passign 'w+' lets us read and write to the file
my_file=open('test.txt','w+')
```

```python
#Write to the file
my_file.write('This is a new first line')
```

```python
#Read the file
my_file.seek(0)
my_file.read()
```

```python
my_file.close()
```

#### Appending to a file

```python
my_file=open('test1.txt','a+')
my_file.write('\nThis line is being appended to a test.txt')
my_file.write('\nAnd another line here.')
```

```python
my_file.seek(0)
print(my_file.read())
```

### Iterating through a File

```python
with open('test1.txt','r') as txt:
    for line in txt:
        print(line, end='')
```

```python
with open('test1.txt','r') as txt:
    for line in txt:
        print(line)
```
