### Pattern

```python
text2="My telephone number is 408-555-1234"
phone=re.search(r'\d\d\d-\d\d\d\-\d\d\d\d',text2)
```

```python
phone.group()
```

####  Groups

```python
phone_pattern=re.compile(r'(\d{3})-(\d{3})-(\d{4})')
results=re.search(phone_pattern,text2)
```

```python
results.group()
```

```python
results.group(1)
```

```python
results.group(2)
```

### The WildCard Character

```python
re.findall(r".at","The cat in te hat sat here.")
```

```python
re.findall(r".at","The bat went splat")
```

```python
re.findall(r"...at","The bat went splat.")
```

```python
#One or more non-whitespace that ends with'at'
re.findall(r'\S+at',"The bat went splat")

#here it is giving the whole word that ends with "at"
```

## Starts With and Ends with

```python
#Ends with a numbert
re.findall(r'\d$','This ends with a number 2')
```

```python
#Starts with a number
re.findall(r'^\d','1 is the lonliest number.')
```

## Exclusion

#### to exclude characyers, we can use the ^ symbol in conjuction with a set of brackets[].
#### anything inside the brackets is excluded

```python
phrase="There are 3 numbers 34 inside 5 in this sentence."
re.findall(r'[^\d]',phrase)
```

```python
re.findall(r'[^\d]+',phrase)
```

```python
test_phrase='This is a string! But it has punctuation. How can we remove it?'
re.findall('[^!.? ]+',test_phrase)
```

```python
# what if we remove space from [] 
re.findall('[^!.?]+',test_phrase)
```

```python
clean=' '.join(re.findall('[^!.? ]+',test_phrase))
```
