```python
doc=nlp(u'This is the first sentence. This is another sentence. This is the last sentece.')

for sent in doc.sents:
    print(sent)
```

## Adding Rules

```python
#Parsing the segmentation syart tokens happens during the nlp pipeline
doc2= nlp(u'This is a sentence. This is a sentence. This i a sentence.')

for token in doc2:
    print(token.is_sent_start, ' '+token.text)
```

```python
doc3=nlp(u'"Management is doing things right; leadership is doing the right things." -Peter Drucker')

for sent in doc3.sents:
    print(sent)
```

```python
#Add a new rule to thecustom pipeling
from spacy.language import Language

@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text ==';':
            doc[token.i+1].is_sent_start =True
    return doc

nlp.add_pipe("set_custom_boundaries", before='parser')

nlp.pipe_names
```
